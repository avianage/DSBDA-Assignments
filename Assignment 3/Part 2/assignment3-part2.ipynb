#### Importing Libraries

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
virus = pd.read_csv("assignment3-part2.csv")
```

```python
virus.replace("?", np.nan, inplace=True)
```

```python
virus.isna().sum()
```

#### Binning

```python
num_bin = 5

label = ["A", "B", "C","D","E"]
```

```python
virus["E-W-Partition"] = pd.cut(virus["V1"], bins=num_bin, labels=label)
virus["E-F-Partition"] = pd.qcut(virus["V1"], q = num_bin, labels=label)
```

```python
virus.iloc[:]
```

#### Dividing Dataset with V1 less than and greater than 40

```python
threshold = 40

virus_low = virus[virus["V1"] < threshold]
virus_high = virus[virus["V1"] > threshold]
```

```python
virus_low.describe()
```

```python
virus_high.describe()
```

#### Plotting Boxplot from V1 for both low and high groups

```python
fig = plt.figure(figsize= (10,5))
```

```python
plt.boxplot(virus_low["V1"])
```

```python
plt.boxplot(virus_high["V1"])
```

```python
virus_low["ANTI_A"] = (virus_low["ANTI_A"] - virus_low["ANTI_A"].min())/(virus_low["ANTI_A"].max() - virus_low["ANTI_A"].min())
virus_low["ANTI_B"] = (virus_low["ANTI_B"] - virus_low["ANTI_B"].min())/(virus_low["ANTI_B"].max() - virus_low["ANTI_B"].min())
virus_low
```

#### Min-Max Normalization

```python
virus_high["ANTI_A"] = (virus_high["ANTI_A"] - virus_high["ANTI_A"].min())/(virus_high["ANTI_A"].max() - virus_high["ANTI_A"].min())
virus_high["ANTI_B"] = (virus_high["ANTI_B"] - virus_high["ANTI_B"].min())/(virus_high["ANTI_B"].max() - virus_high["ANTI_B"].min())
virus_high
```

#### Plotting Scatterplot

```python
plt.scatter(virus_low["ANTI_A"], virus_low["V1"], c="blue")
plt.xlabel("ANTI_A")
plt.ylabel("V1")
plt.show()
```

```python
plt.scatter(virus_low["ANTI_B"], virus_low["V1"], c="blue")
plt.xlabel("ANTI_B")
plt.ylabel("V1")
plt.show()
```

```python
plt.scatter(virus_high["ANTI_A"], virus_high["V1"], c="red")
plt.xlabel("ANTI_A")
plt.ylabel("V1")
plt.show()
```

```python
plt.scatter(virus_high["ANTI_B"], virus_high["V1"], c="red")
plt.xlabel("ANTI_B")
plt.ylabel("V1")
plt.show()
```

#### Corelation between ANTI_A and ANTI_B

```python
virus_low["ANTI_A"].corr(virus_low["ANTI_B"])
```

```python
virus_high["ANTI_A"].corr(virus_high["ANTI_B"])
```

For low values, we must use only one of the medicines i.e ANTI_A or ANTI_B as the corelation is negative but for high values, we can use both ANTI_A and ANTI_B together as the corelation is positive
